### Importação de Bibliotecas

```python
import pandas as pd
import pathlib
import win32com.client as win32
```

### Importação e Leitura das Bases de Dados

```python
df_emails = pd.read_excel(r'Bases de Dados/Emails.xlsx')
df_lojas = pd.read_csv(r'Bases de Dados/Lojas.csv', sep=';', encoding='latin')
df_vendas = pd.read_excel(r'Bases de Dados/Vendas.xlsx')

display(df_emails.head())
display(df_lojas.head())
display(df_vendas.head())
```

```python
print(df_vendas.info())
print(df_lojas.info())
```

### Criação de um dataframe que mescle as informações de vendas e lojas

```python
# Cria o df mesclando as informações de vendas e lojas
df_vendas_lojas = df_vendas.merge(df_lojas, on='ID Loja')
print(df_vendas_lojas.info())
```

### Salvar planilha de cada loja na pasta de backup

```python
# cria um dataframe para armazenar {'nome_loja': df_da_loja}

dicionario_info_lojas = {} 

# preenche o dicionario 

for loja in df_lojas['Loja']: # percorre o a coluna 'Loja' do df das lojas
    dicionario_info_lojas[loja] = df_vendas_lojas.loc[df_vendas_lojas['Loja']==loja,:]
    ''' cria um item no dicionario com o nome da loja e o df com as vendas referentes a loja 
       a partir completo(mesclado) de vendas e lojas '''
display(dicionario_info_lojas['Iguatemi Esplanada'])
```

### Definir o dia do indicador

```python
dia_calculo_indicador = df_vendas_lojas['Data'].max() # mantém o cálculo do indicador atualizado

print(dia_calculo_indicador)
```

### Criar uma tabela para cada loja

```python
caminho_backup = pathlib.Path(r'Backup Arquivos Lojas') # coloca em uma variavel o caminho a ser percorrido

arquivos_pasta_backup = caminho_backup.iterdir() # pega todos os arquivos que tem no caminho

lista_nome_arquivos = [arquivo.name for arquivo in arquivos_pasta_backup] # cria uma lista com o nome dos arquivos

for loja in dicionario_info_lojas: # verifica se a pasta existe, cria se nao existir, e armazena os arquivos em suas respectivas pastas
    if loja not in lista_nome_arquivos: # verifica se NAO existe a pasta referente a loja
        nova_pasta = caminho_backup/loja # define a localizacao e nome da pasta 
        nova_pasta.mkdir() # cria a pasta

    nome_arquivo = f'{dia_calculo_indicador.month}_{dia_calculo_indicador.day}_{loja}.xlsx' # nome do arquivo

    local_armazen_arquivo = caminho_backup / loja / nome_arquivo # variavel que armazena o caminho completo do ARQUIVO

    dicionario_info_lojas[loja].to_excel(local_armazen_arquivo) #gera o arquivo excel com cada dataframe do dicionario.
```

### Definição das Metas Pré Estabelecidas

```python
#definição de metas

meta_faturamento_dia = 1000
meta_faturamento_ano = 1650000
meta_qtdeprodutos_dia = 4
meta_qtdeprodutos_ano = 120
meta_ticketmedio_dia = 500
meta_ticketmedio_ano = 500
```

### Cálculo de indicadores e disparo de e-mail para cada loja da rede.

```python
for loja in dicionario_info_lojas:

    # cria-se um dataframe auxiliar filtrando pela data para o calculo dos indicadores diários

    df_aux_vendas_lojas_dia = dicionario_info_lojas[loja].loc[dicionario_info_lojas[loja]['Data']==dia_calculo_indicador, :]
    # faturamento - somar valor das vendas de cada loja
    faturamento_ano = dicionario_info_lojas[loja]['Valor Final'].sum()
    faturamento_dia = df_aux_vendas_lojas_dia['Valor Final'].sum()

    # diversidade de produtos vendidos:
    div_prod_vend_ano = dicionario_info_lojas[loja]['Produto'].nunique() #conta o numero de valores unicos
    div_prod_vend_dia = df_aux_vendas_lojas_dia['Produto'].nunique()

    # ticket medio por venda:
    # criam-se df auxiliares com agrupamento por codigo de venda para o calculo do ticket medio por venda.
    df_aux_agroup_por_vendas_ano = dicionario_info_lojas[loja].groupby('Código Venda').sum(numeric_only=True).sort_values(by='Código Venda')
    df_aux_agroup_por_vendas_dia = df_aux_vendas_lojas_dia.groupby('Código Venda').sum(numeric_only=True).sort_values(by='Código Venda')

    tkt_med_ano = df_aux_agroup_por_vendas_ano['Valor Final'].mean()
    tkt_med_dia = df_aux_vendas_lojas_dia['Valor Final'].mean()

        #enviar o e-mail
    outlook = win32.Dispatch('outlook.application')

    nome = df_emails.loc[df_emails['Loja']==loja, 'Gerente'].values[0]
    mail = outlook.CreateItem(0)
    mail.To = df_emails.loc[df_emails['Loja']==loja, 'E-mail'].values[0]
    mail.Subject = f'OnePage Dia {dia_calculo_indicador.day}/{dia_calculo_indicador.month} - Loja {loja}'
    #mail.body = 'Texto do E-mail'

    if faturamento_dia >= meta_faturamento_dia:
            cor_fat_dia = 'green'
    else:
        cor_fat_dia = 'red'
    if faturamento_ano >= meta_faturamento_ano:
        cor_fat_ano = 'green'
    else:
        cor_fat_ano = 'red'
    if div_prod_vend_dia >= meta_qtdeprodutos_dia:
        cor_qtde_dia = 'green'
    else:
        cor_qtde_dia = 'red'
    if div_prod_vend_ano >= meta_qtdeprodutos_ano:
        cor_qtde_ano = 'green'
    else:
        cor_qtde_ano = 'red'
    if tkt_med_dia >= meta_ticketmedio_dia:
        cor_ticket_dia = 'green'
    else:
        cor_ticket_dia = 'red'
    if tkt_med_ano >= meta_ticketmedio_ano:
        cor_ticket_ano = 'green'
    else:
        cor_ticket_ano = 'red'

    mail.HTMLBody = f'''
    <p>Bom dia, {nome}</p>

    <p>O resultado de ontem <strong>({dia_calculo_indicador.day}/{dia_calculo_indicador.month})</strong> da <strong>Loja {loja}</strong> foi:</p>

    <table>
        <tr>
        <th>Indicador</th>
        <th>Valor Dia</th>
        <th>Meta Dia</th>
        <th>Cenário Dia</th>
        </tr>
        <tr>
        <td>Faturamento</td>
        <td style="text-align: center">R${faturamento_dia:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_dia:.2f}</td>
        <td style="text-align: center"><font color="{cor_fat_dia}">◙</font></td>
        </tr>
        <tr>
        <td>Diversidade de Produtos</td>
        <td style="text-align: center">{div_prod_vend_dia}</td>
        <td style="text-align: center">{meta_qtdeprodutos_dia}</td>
        <td style="text-align: center"><font color="{cor_qtde_dia}">◙</font></td>
        </tr>
        <tr>
        <td>Ticket Médio</td>
        <td style="text-align: center">R${tkt_med_dia:.2f}</td>
        <td style="text-align: center">R${meta_ticketmedio_dia:.2f}</td>
        <td style="text-align: center"><font color="{cor_ticket_dia}">◙</font></td>
        </tr>
    </table>
    <br>
    <table>
        <tr>
        <th>Indicador</th>
        <th>Valor Ano</th>
        <th>Meta Ano</th>
        <th>Cenário Ano</th>
        </tr>
        <tr>
        <td>Faturamento</td>
        <td style="text-align: center">R${faturamento_ano:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_ano:.2f}</td>
        <td style="text-align: center"><font color="{cor_fat_ano}">◙</font></td>
        </tr>
        <tr>
        <td>Diversidade de Produtos</td>
        <td style="text-align: center">{div_prod_vend_ano}</td>
        <td style="text-align: center">{meta_qtdeprodutos_ano}</td>
        <td style="text-align: center"><font color="{cor_qtde_ano}">◙</font></td>
        </tr>
        <tr>
        <td>Ticket Médio</td>
        <td style="text-align: center">R${tkt_med_ano:.2f}</td>
        <td style="text-align: center">R${meta_ticketmedio_ano:.2f}</td>
        <td style="text-align: center"><font color="{cor_ticket_ano}">◙</font></td>
        </tr>
    </table>

    <p>Segue em anexo a planilha com todos os dados para mais detalhes.</p>

    <p>Qualquer dúvida estou à disposição.</p>
    <p>Att., Igor</p>
    '''


    # Anexos (pode colocar quantos quiser):
    attachment  = pathlib.Path.cwd() / caminho_backup / loja / f'{dia_calculo_indicador.month}_{dia_calculo_indicador.day}_{loja}.xlsx'
    mail.Attachments.Add(str(attachment))

    mail.Send()
    

print('E-mails enviados')
```

### Criação de Ranking para a Diretoria

```python
df_faturamento_anual_loja = df_vendas_lojas.groupby('Loja').sum(numeric_only=True).sort_values(by='Valor Final', ascending=False)
df_faturamento_diario_loja = df_vendas_lojas.loc[df_vendas_lojas['Data']==dia_calculo_indicador,:].groupby('Loja').sum(numeric_only=True).sort_values(by='Valor Final', ascending=False)
display(df_faturamento_anual_loja)
display(df_faturamento_diario_loja)
```

```python
df_faturamento_anual_loja = df_faturamento_anual_loja.reset_index()
df_faturamento_anual_loja = df_faturamento_anual_loja.loc[:,['Loja','Valor Final']]

nome_arquivo = f'{dia_calculo_indicador.month}_{dia_calculo_indicador.day}_Ranking Anual.xlsx' # nome do arquivo

local_armazen_arquivo = caminho_backup / nome_arquivo # variavel que armazena o caminho completo do ARQUIVO

df_faturamento_anual_loja.to_excel(local_armazen_arquivo) 

df_faturamento_diario_loja = df_faturamento_diario_loja.reset_index()
df_faturamento_diario_loja = df_faturamento_diario_loja.loc[:,['Loja','Valor Final']]

nome_arquivo = f'{dia_calculo_indicador.month}_{dia_calculo_indicador.day}_Ranking Diario.xlsx' # nome do arquivo

local_armazen_arquivo = caminho_backup / nome_arquivo # variavel que armazena o caminho completo do ARQUIVO

df_faturamento_diario_loja.to_excel(local_armazen_arquivo) #transforma o df em um arquivo excel e armazena no caminho definido

display(df_faturamento_anual_loja)
display(df_faturamento_diario_loja)
```
